```python
%matplotlib inline
import pandas as pd
import mplfinance as mpf
file = 'GBPJPY_H1_updated.csv'
df = pd.read_csv(file)


#splitting date column and drop unnamed column
df[['Date', 'Time']] = df['Date'].str.split(' ', 1, expand=True)

#create a table with new column (Time)
df = df[['Date','Time','Open', 'High', 'Low','Close','Volume']]
df
```

```python
#start_date = '2023-05-01'
#end_date = '2023-09-10'
#x = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
#x

#get certain date



def session_visualize(when):
    certain_date = df[df['Date'] == when]
    certain_date
    
    #get certain time
    start_time = '03:00'
    end_time = '11:00'
    certain_datetime = certain_date[(certain_date['Time'] >= start_time) & (certain_date['Time'] <= end_time)]
    certain_datetime
    
    #candlestick
    certain_datetime['Time'] = pd.to_datetime(certain_datetime['Time'])  # Make sure 'Date' column is in datetime format
    certain_datetime.set_index('Time', inplace=True) # Set the 'Date' column as index
    mpf.plot(certain_datetime, type='candle', style='classic', title=when, ylabel='level') # Create a candlestick chart
    
def whole_day_visualize(when):
    certain_date = df[df['Date'] == when]
    certain_date
    
    #get certain time
    start_time = '00:00'
    end_time = '21:00'
    certain_datetime = certain_date[(certain_date['Time'] >= start_time) & (certain_date['Time'] <= end_time)]
    certain_datetime
    
    #candlestick
    certain_datetime['Time'] = pd.to_datetime(certain_datetime['Time'])  # Make sure 'Date' column is in datetime format
    certain_datetime.set_index('Time', inplace=True) # Set the 'Date' column as index
    mpf.plot(certain_datetime, type='candle', style='classic', title=when, ylabel='level') # Create a candlestick chart
```
